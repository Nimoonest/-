```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime, timedelta

# 生成时间序列：从10:00到12:00，每隔30分钟
start_time = datetime.strptime("10:00", "%H:%M")
time_list = [start_time + timedelta(minutes=30*i) for i in range(5)]  # 10:00, 10:30, ..., 12:00

# 示例数据
y = [1, 3, 2, 4, 5]

# 创建图
plt.figure(figsize=(8, 4))
plt.plot(time_list, y, marker='o')

# 设置x轴格式
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))  # 设置时间显示格式为10:30
plt.gca().xaxis.set_major_locator(mdates.MinuteLocator(interval=30))  # 每30分钟一个刻度

plt.xlabel("time")
plt.ylabel("value")
plt.title('U')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import random
x = range(0,120)
a = [random.randint(20,35) for i in range(120)]
plt.figure(figsize=(30, 6))
plt.plot(x, a)
_xtick_labels = ['hello,{}'.format(i) for i in x] # 输入字符串
plt.xticks(ticks = x[::10],labels = _xtick_labels[::10], rotation = 45) # ticks：刻度位置 labels:刻度内容 rotation:旋转角度
'''sequence[start:stop:step] start：起始索引（默认从开头）
stop：终止索引（不包括这个位置，默认到结尾）
step：步长（每隔多少取一个）'''
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [2, 4, 5, 7, 9]
y = [4, 5, 7, 9, 4]
```

```python
plt.plot(x,y)
plt.title('1')
plt.xlabel('axis X')
plt.ylabel('axis Y')
plt.show()
```

```python
plt.figure(figsize=(6, 6))  # 单位是英寸
#设置坐标轴范围
plt.xlim(0, 10)  # x轴范围
plt.xticks(range(0,11)) # 设置x轴刻度
plt.ylim(3, 10) # y轴范围
plt.plot(x, y, color='blue', linestyle='--', linewidth=1, marker='o') 
'''color='red'：线的颜色（可用英文名或 #RRGGBB 十六进制）
linestyle='--'：虚线风格（常见的有 '-', '--', ':', '-.'）
linewidth=1：线条宽度
marker='o'：在每个点上加一个圆圈标记（常见有 'o', '^', 's', 'x' 等）'''
plt.annotate('highest', xy=(7, 9), xytext=(6,9.5),
             arrowprops=dict(facecolor='black', shrink=0.05)) # 标记特殊点
plt.plot(2, 4, 'ro')  # 用红点标记
plt.text(2, 4, ' (2,4)', fontsize=12)  # 加文字
plt.text(0.5, 0.5, 'Confidential', fontsize=40,
         color='gray', alpha=0.3, ha='center', va='center',
         transform=plt.gca().transAxes)
'''transform=plt.gca().transAxes：确保水印按图大小比例定位（而不是数据值）
alpha=0.3：设置透明度
ha/va='center'：水平、垂直居中'''
# plt.savefig("图表.png")  #保存图片 默认当前路径，也可以手动输入指定
plt.show()
```
