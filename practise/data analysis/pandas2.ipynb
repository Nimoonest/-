```python
import numpy as np
import pandas as pd
```

```python
dates = pd.date_range('20130101', periods = 6)
df = pd.DataFrame(np.arange(24).reshape((6, 4)), index = dates, columns = ['A', 'B', 'C', 'D'])
df
```

```python
print(df['A'],df.A) # 打印某列
```

```python
print(df[0:3], '\n',df[::2]) # 打印行 切片
```

```python
# select by label:loc  根据标签选择
print(df.loc['20130102'])
```

```python
print(df.loc[:, 'A']) # 保存所有行的数据并对列的数据进行筛选
```

```python
print(df.loc['20130102',['A', 'B']]) # 可以同时根据行、列标签进行筛选
```

```python
# select by position: iloc  根据位置筛选
print(df.iloc[3,1])
```

```python
print(df.iloc[1:3, 1:3])  # 筛选第2、3列，第2、3行
```

```python
print(df.iloc[[1, 3, 5], 1:3]) # 不连续筛选 第二、四、六行，第二到三列
```

```python
# Boolean indexing  布尔型筛选
print(df)
print(df[df.A > 8])
```

```python
import numpy as np
import pandas as pd
dates= pd.date_range('20130101',periods = 6)
df = pd.DataFrame(np.arange(24).reshape(6,4), index = dates, columns = ['A', 'B', 'C', 'D'])
df
```

```python
df.iloc[2, 2] = 1111
df
```

```python
df.loc['20130101', 'B'] = 2222
df
```

```python
df[df.A > 4] = 0  # 使A变量大于4的行等于0
df
```

```python
df.loc[df.A > 4, 'A'] = 0
df
```

```python
dates= pd.date_range('20130101',periods = 6)
df = pd.DataFrame(np.arange(24).reshape(6,4), index = dates, columns = ['A', 'B', 'C', 'D'])
df
# 筛选 A > 4 且 B < 10 的行，把 C 列赋值为 999
df.loc[(df['A'] > 4) & (df['B'] < 10), 'C'] = 999
df
```

```python
# 筛选 A > 10 或 D < 5 的行，把 B 列设为 0
df.loc[(df['A'] > 10) | (df['D'] < 5), 'B'] = 0
df
```

```python
df['F'] = np.nan
df
```

```python
df.loc[:,'F'] = 1
df
```

```python
df['E'] = pd.Series([1, 2, 3, 4, 5, 6],index = pd.date_range('20130101',periods = 6)) # 添加E列，行标签配齐
df
```

```python
df['G'] = pd.Series(range(5,11),index = df.index)  #添加G列
df
```

```python
dates= pd.date_range('20130101',periods = 6)
df = pd.DataFrame(np.arange(24).reshape(6,4), index = dates, columns = ['A', 'B', 'C', 'D'])
df.iloc[0, 1] = np.nan
df.iloc[1, 2] = np.nan
df
```

```python
print(df.dropna(axis = 0, how = 'any')) # how = 'any' 有nan就隐藏掉   how = {'any','all'}  all全是nan才隐藏掉  
# axis = 0, 向左方向索引  隐藏行
```

```python
print(df.dropna(axis = 1, how = 'any')) # axis = 1, 向右方向索引  隐藏列
```

```python
print(df.fillna(value = 0)) # 为nan填入值
```

```python
print(df.isnull())   # 输出是否为null
```

```python
print(np.any(df.isnull()) == True)  # 检测有无null值
```
