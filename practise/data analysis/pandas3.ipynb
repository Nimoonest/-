| 文件类型    | 函数                           | 示例                                  |
| ------- | ---------------------------- | ----------------------------------- |
| CSV     | `pd.read_csv()`              | `pd.read_csv('data.csv')`           |
| TSV/制表符 | `pd.read_csv(..., sep='\t')` | `pd.read_csv('data.tsv', sep='\t')` |
| TXT     | `pd.read_table()`            | `pd.read_table('data.txt')`         |


| 类型        | 函数                    | 示例                                         |
| --------- | --------------------- | ------------------------------------------ |
| JSON      | `pd.read_json()`      | `pd.read_json('data.json')`                |
| HTML 表格   | `pd.read_html()`      | `pd.read_html('https://...')`              |
| SQL 数据库   | `pd.read_sql()`       | `pd.read_sql('SELECT * FROM table', conn)` |
| Clipboard | `pd.read_clipboard()` | `pd.read_clipboard()`                      |
| Parquet   | `pd.read_parquet()`   | `pd.read_parquet('data.parquet')`          |
| Pickle    | `pd.read_pickle()`    | `pd.read_pickle('data.pkl')`               |
