```python
df1 = pd.DataFrame(np.ones((3,4))*0, columns = ['a', 'b', 'c', 'd'], index = [1, 2, 3])
df2 = pd.DataFrame(np.ones((3,4))*1, columns = ['b', 'c', 'd', 'e'], index = [2, 3, 4])
df1
```

```python
df2
```

```python
res = pd.concat([df1, df2], axis = 0, join = 'inner')  # 只保留公共索引
res
```

```python
res = pd.concat([df1, df2], axis = 0, join = 'outer')  # 保留所有索引
res
```

```python
res = pd.concat([df1, df2], axis = 0, join = 'inner', ignore_index = True)  # 只保留公共索引
res
```

```python
# keys 用于生成多级索引
df1 = pd.DataFrame(np.ones((3,4))*0, columns = ['a', 'b', 'c', 'd'], index = [1, 2, 3])
df2 = pd.DataFrame(np.ones((3,4))*1, columns = ['b', 'c', 'd', 'e'], index = [2, 3, 4])
res = pd.concat([df1, df2], axis = 1, keys = ['df1','df2'])
res
```

verify_integrity	如果为 True，会检查索引是否重复，重复就报错

| 参数                     | 说明                                          |
| ---------------------- | ------------------------------------------- |
| `left` / `right`       | 要合并的两个 DataFrame                            |
| `how`                  | 合并方式：`'inner'`、`'outer'`、`'left'`、`'right'` |
| `on`                   | 用于连接的列名（两边有相同列时可用）                          |
| `left_on` / `right_on` | 如果连接列在两边名字不同，用这两个指定                         |


| `how`     | 类似 SQL            | 说明                |
| --------- | ----------------- | ----------------- |
| `'inner'` | `INNER JOIN`      | 只保留两边都存在的键（交集）    |
| `'outer'` | `FULL OUTER JOIN` | 保留所有键（并集），缺的补 NaN |
| `'left'`  | `LEFT JOIN`       | 以左边为主，右边补齐        |
| `'right'` | `RIGHT JOIN`      | 以右边为主，左边补齐        |


```python
left = pd.DataFrame({'key':['K0', 'K1', 'K2', 'K3'],
                    'A':['A0', 'A1', 'A2','A3'],
                    'B':['B0', 'B1', 'B2','B3']})
right = pd.DataFrame({'key':['K0', 'K1', 'K2', 'K3'],
                    'C':['C0', 'C1', 'C2','C3'],
                    'D':['D0', 'D1', 'D2','D3']})
```

```python
left
```

```python
right
```

```python
df = pd.merge(left, right, on='key', how='inner')
df
```

```python
# two keys
left = pd.DataFrame({'key1':['K0', 'K0', 'K1', 'K2'],
                     'key2':['K0', 'K1', 'K0', 'K1'],
                    'A':['A0', 'A1', 'A2','A3'],
                    'B':['B0', 'B1', 'B2','B3']})
right = pd.DataFrame({'key1':['K0', 'K1', 'K1', 'K2'],
                      'key2':['K0', 'K0', 'K0', 'K0'],
                    'C':['C0', 'C1', 'C2','C3'],
                    'D':['D0', 'D1', 'D2','D3']})
left
```

```python
right
```

```python
res = pd.merge(left, right, on = ['key1', 'key2'])
res
```

```python
res = pd.merge(left, right, on = ['key1', 'key2'], how = 'outer')
res
```

```python
res = pd.merge(left, right, on = ['key1', 'key2'], how = 'left', indicator = True)  #用于显示每一行来自哪个 DataFrame
res
```

```python
left = pd.DataFrame({'A':['A0', 'A1', 'A2'],
                    'B':['B0', 'B1', 'B2']},
                    index = ['K0', 'K1', 'K2'])
right = pd.DataFrame({'C':['C0', 'C2', 'C3'],
                    'D':['D0', 'D2', 'D3']},
                     index = ['K0', 'K2', 'K3'])
left
```

```python
right
```

```python
res = pd.merge(left, right, left_index = True, right_index = True)
res
```

```python
res = pd.merge(left, right, left_index = True, right_index = True,how = 'outer')
res
```

pd.merge(df1, df2, on='key', suffixes=('_左表', '_右表')) 用于在两张表中存在相同列名时，自动为这些重复列加上“后缀”

```python
left = pd.DataFrame({'key':['K0', 'K1', 'K2', 'K3'],
                    'A':['A0', 'A1', 'A2','A3'],
                    'B':['B0', 'B1', 'B2','B3']})
right = pd.DataFrame({'key':['K0', 'K1', 'K2', 'K3'],
                    'A':['C0', 'C1', 'C2','C3'],
                    'D':['D0', 'D1', 'D2','D3']})
res = pd.merge(left, right, on='key', suffixes=('_left', '_right'))
res
```
